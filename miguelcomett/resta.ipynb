```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar las imágenes
low_energy_img = cv2.imread('Results/arm30kev.jpg', cv2.IMREAD_GRAYSCALE)
high_energy_img = cv2.imread('Results/arm70kev.jpg', cv2.IMREAD_GRAYSCALE)

def sls(low_energy_img, high_energy_img, w):
    # Convertir imágenes a float para evitar errores de precisión
    low_energy_img = np.float64(low_energy_img)
    high_energy_img = np.float64(high_energy_img)

    # Aplicar la sustracción ponderada
    tissue_img = -w * low_energy_img + high_energy_img
    bone_img = w * low_energy_img - high_energy_img

    return tissue_img, bone_img

# Factor de ponderación w
w = 1
tissue_img, bone_img = sls(low_energy_img, high_energy_img, w)



# Muestra las imágenes antes y después de la difusión
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].imshow(low_energy_img,cmap = 'gray')
ax[0, 0].set_title("Imagen 1 Original")
ax[0, 1].imshow(high_energy_img,cmap = 'gray')
ax[0, 1].set_title("Imagen 2 Original")
ax[1, 0].imshow(tissue_img,cmap = 'gray')
ax[1, 0].set_title("Tissue Image")
ax[1, 1].imshow(bone_img,cmap = 'gray')
ax[1, 1].set_title("Bone Imagwe")

plt.show()
```
