```python
import os
import pygetwindow as gw
import subprocess
import time

# Ruta del ejecutable Sim.exe
sim_path = r'C:\Users\conej\Documents\Universidad\Geant4\Projects\Main\MedX_Engine\build\Release\Sim.exe'

# Función para ejecutar Sim.exe desde el directorio especificado
def run_simulation():
    # Cambiar al directorio de Sim.exe
    sim_dir = os.path.dirname(sim_path)
    os.chdir(sim_dir)
    
    # Ejecutar Sim.exe
    process = subprocess.Popen([sim_path])  
    print(f"Sim.exe (PID: {process.pid}) se está ejecutando...")
    return process

# Función para obtener y mostrar las ventanas abiertas
def list_open_windows():
    windows = gw.getWindowsWithTitle('')  # Listar todas las ventanas abiertas
    print("Ventanas abiertas:")
    for window in windows:
        print(f"- {window.title}")  # Imprimir el título de cada ventana
    return windows

def main():
    # Ejecutar Sim.exe
    process = run_simulation()
    
    # Esperar unos segundos para que las ventanas se abran
    time.sleep(5)
    
    # Obtener y listar las ventanas abiertas
    windows = list_open_windows()
    
    # Asegurarse de que el proceso siga ejecutándose
    while process.poll() is None:
        time.sleep(2)  # Esperar un poco antes de obtener las ventanas nuevamente
        # Obtener y listar ventanas nuevamente si es necesario
        windows = list_open_windows()

    print("Sim.exe ha terminado.")

if __name__ == "__main__":
    main()
```
