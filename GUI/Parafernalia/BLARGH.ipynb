```python
import pandas as pd
```

```python
df = pd.read_csv('Etiquetados_evento_ZMM.csv')
df.head()
```

```python
# Tomar una muestra aleatoria de 1000 caid Ãºnicos
sampled_caids = df['caid'].drop_duplicates().sample(n=1000, random_state=42)

# Filtrar el DataFrame para obtener todas las filas con esos caid seleccionados
sampled_df = df[df['caid'].isin(sampled_caids)]

# Mostrar las primeras filas de la muestra
print(sampled_df.head())
```

```python
sampled_df.head()
```

```python
# Guardar el DataFrame en un archivo CSV
sampled_df.to_csv('muestra_1000_caid.csv', index=False)

print("El archivo ha sido guardado como 'muestra_1000_caid.csv'")
```
